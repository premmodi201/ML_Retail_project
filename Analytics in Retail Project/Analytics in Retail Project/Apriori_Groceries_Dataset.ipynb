```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('/content/drive/MyDrive/Groceries_dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.itemDescription.unique()
```

```python
df['itemDescription'] = df['itemDescription'].str.strip()
df.dropna(axis=0, subset=['Member_number'], inplace=True)
df['Member_number'] = df['Member_number'].astype('str')
df = df[~df['Member_number'].str.contains('C')]
```

```python
def hot_encode(x):
	if(x<= 0):
		return 0
	if(x>= 1):
		return 1
```

```python
basket_encoded = df.groupby(['Member_number', 'itemDescription'])['itemDescription'].count().unstack().reset_index().fillna(0).set_index('Member_number')
basket_encoded = basket_encoded.applymap(hot_encode)
basket_encoded
```

```python
basket_encoded.applymap(hot_encode)
basket_encoded
```

```python
frq_items = apriori(basket_encoded, min_support = 0.05, use_colnames = True, verbose=0)
frq_items
```

```python
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules
```
